```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver as wd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common import exceptions
import time
import re
import pandas as pd
import random


keyword = "에슬레져룩"

url1 = "https://www.instagram.com"
url = "https://www.instagram.com/explore/tags/{}/".format(keyword)

instagram_tags = []
instagram_tag_dates = []

driver = wd.Chrome("./chromedriver")
driver.get(url1)
time.sleep(3)

inputID = driver.find_element_by_name("username")
inputID.send_keys("id")
inputPW = driver.find_element_by_name("password")
inputPW.send_keys("pw")

driver.find_element_by_class_name('sqdOP.L3NKy.y3zKF').click()
time.sleep(3)

driver.find_element_by_class_name('sqdOP.yWX7d.y3zKF').click()   
time.sleep(3)

driver.find_element_by_class_name('aOOlW.HoLwm').click()
time.sleep(3)

driver.get(url)
time.sleep(5)

driver.find_element_by_class_name('_9AhH0').click()
for i in range(500):
    time.sleep(random.uniform(3,5))
    try:
        data = driver.find_element_by_class_name('C4VMK')
        tag_raw = data.text
        tags = re.findall('#[A-Za-z0-9가-힣]+', tag_raw)
        tag = ''.join(tags).replace("#"," ")
        tag_data = tag.split()
        
        for tag_one in tag_data:
            instagram_tags.append(tag_one)
            #print(instagram_tags)
            
            
        date = driver.find_element_by_class_name('_1o9PC Nzb55').text
        instagram_tag_dates.append(date)
#         if date.find('시간') != -1 or date.find('일') != -1 or date.find('분') != -1:
#             instagram_tag_dates.append('0주')
#         else: 
#             instagram_tag_dates.append(date)
#             #print(instagram_tag_dates)
        
    except:
        instagram_tags.append("null")
        instagram_tag_dates.append('null')
    try:
        WebDriverWait(driver, 100).until(EC.presence_of_element_located((By.CLASS_NAME, '_65Bje.coreSpriteRightPaginationArrow')))
        driver.find_element_by_class_name('_65Bje.coreSpriteRightPaginationArrow').click()
    except:
        driver.close()
        #date = datum2.text
        #print(date)
    time.sleep(random.uniform(3,5))

final = pd.DataFrame(instagram_tags)
final.to_csv('instagram_result_에슬레져룩.csv', encoding='utf-8-sig')

driver.close()
driver.quit()
```
